````python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        
class LinkedList:
    def __init__(self):
        self.head = None
        
    def append(self, item):
        
        new_node = Node(item)
        if not self.head:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node
            
    def insert(self, index, item):  # this method should insert a new node with "item" for data BEFORE the target index.
        new_node = Node(item)
        if not self.head:
            self.head = new_node
        elif index == 0:
            new_node.next = self.head
            self.head = new_node
        else:
            counter = 0
            current = self.head
            prev = None
            while current.next and index != counter: # while current.next != 1...
                prev = current
                current = current.next
                counter += 1
            if index == counter
                prev.next = new_node
                new_node.next = current
            else:
                current.next = new_node
                
                    #remove(self, value, /)
                    #Remove first occurrence of value.
                    #Raises ValueError if the value is not present.   
    def remove(self, value):
        current = self.head
        if not current:
            raise ValueError("There's nothing to remove!")
        if current.data == value:
            self.head = current.next
        
        else:
            prev = None
            while current.next and current.data != value
                prev = current
                current = current.next
            if current.data == value:
                prev.next = current.next
            else: 
                raise ValueError("%s is not in list>" % value)
                
                
                
````

````python
mylist = LinkedList()
for i in range(5):
    mylist.append(i)

current = mylist.head
for i in range(5):
    assert current.data == i
    current = current.next

mylist.insert(0, -1)

assert mylist.head.data == -1
````
